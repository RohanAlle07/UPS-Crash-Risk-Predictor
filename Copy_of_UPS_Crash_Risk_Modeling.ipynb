***Handle Date/Time***

```python
# Preserve original (always good practice)
df['report_date_raw'] = df['report_date']
df['report_time_raw'] = df['report_time']


# --- Fix report_date ---
df['report_date'] = pd.to_datetime(
    df['report_date'].astype(str),
    format='%Y%m%d',
    errors='coerce'
)


# --- Fix report_time ---
# Ensure 4-digit format (e.g., 30 -> "0030")
df['report_time'] = df['report_time'].astype(str).str.zfill(4)

df['report_time'] = pd.to_datetime(
    df['report_time'],
    format='%H%M',
    errors='coerce'
).dt.time
```

***Simplify Weather***

Group logically

```python
def simplify_weather(x):
    if x in ['no adverse conditions']:
        return 'clear'
    elif x in ['rain', 'sleet hail']:
        return 'rain'
    elif x in ['snow', 'blowing sand soil dirt or snow']:
        return 'snow'
    elif x in ['fog']:
        return 'fog'
    elif x in ['severe crosswinds']:
        return 'wind'
    else:
        return 'other'

df1['weather_simple'] = df1['weather_condition_desc'].apply(simplify_weather)
```

```python
def simplify_light(x):
    if 'daylight' in x:
        return 'day'
    elif 'dark' in x:
        return 'night'
    elif x in ['dawn', 'dusk']:
        return 'twilight'
    else:
        return 'unknown'

df1['light_simple'] = df1['light_condition_desc'].apply(simplify_light)
```

```python
def simplify_road(x):
    if x == 'dry':
        return 'dry'
    elif x in ['wet', 'water(standing moving)']:
        return 'wet'
    elif x in ['snow', 'slush']:
        return 'snow'
    elif x == 'ice':
        return 'ice'
    else:
        return 'other'

df1['road_simple'] = df1['road_surface_condition_desc'].apply(simplify_road)
```

```python
def simplify_traffic(x):
    if 'not divided' in x:
        return 'undivided'
    elif 'divided' in x:
        return 'divided'
    elif 'one-way' in x:
        return 'one_way'
    else:
        return 'other'

df1['traffic_simple'] = df1['trafficway_desc'].apply(simplify_traffic)
```

# **Format Output for Driver (Readable Risk Report)**

```python
def format_risk_output_v2(row):
    fatal = round(row['Fatal'] * 100, 1)
    injury = round(row['Injury'] * 100, 1)
    pdmg = round(row['Property_Damage'] * 100, 1)

    # Determine highest risk
    risks = {
        "Fatal": fatal,
        "Injury": injury,
        "Minor Damage": pdmg
    }

    main_risk = max(risks, key=risks.get)
    main_value = risks[main_risk]

    print("🚚 DELIVERY RISK SUMMARY")
    print("========================")

    print(f"👉 Primary Risk: {main_risk} ({main_value}%)\n")

    print("Breakdown:")
    print(f" - Fatal: {fatal}%")
    print(f" - Injury: {injury}%")
    print(f" - Minor Damage: {pdmg}%")

    # Action message
    print("\nDriver Guidance:")

    if main_risk == "Fatal":
        print("🔴 HIGH ALERT: High severity risk. Reduce speed and increase following distance.")
    elif main_risk == "Injury":
        print("🟠 CAUTION: Elevated injury risk. Stay alert and avoid aggressive driving.")
    else:
        print("🟢 NORMAL: No major risk indicators, but remain attentive.")
```
